# **Aquisição e Preparação dos dados**








```python
#@markdown `códigos alpha-3`

# Obter todos os códigos ISO-3166 alpha-3 (padrão usado pelo World Bank)
country_codes = [country.alpha_3 for country in pycountry.countries]

# Códigos correspondentes à renda mensal
income_codes = ['HIC', 'LMC', 'UMC']
```

```python
#@markdown `Função de Desagregação`
def split_disaggregation_column(df, prefix=None):
    """
    Divide a coluna 'Disaggregation' de um DataFrame em múltiplas colunas,
    com base em um separador padrão ou prefixo específico.

    Parâmetros:
    - df: pd.DataFrame com a coluna 'Disaggregation'
    - prefix: string usada como separador fixo (ex: 'female, ').
              Se None, será usado ', ' como separador padrão.

    Retorna:
    - df modificado com as novas colunas adicionadas e 'Disaggregation' removida.
    """
    if 'Disaggregation' not in df.columns:
        return df

    if prefix is not None:
        disagg = df['Disaggregation'].str.split(prefix, expand=True)
        if disagg.shape[1] >= 2:
            disagg.rename(columns={1: 'Graduation'}, inplace=True)
            df = pd.concat([df, disagg[['Graduation']]], axis=1)
    else:
        disagg = df['Disaggregation'].str.split(', ', expand=True)
        if disagg.shape[1] >= 2:
            disagg.rename(columns={0: 'Scholarity', 1: 'Gender'}, inplace=True)
            df = pd.concat([df, disagg[['Scholarity', 'Gender']]], axis=1)

    df = df.drop(columns=['Disaggregation'], errors='ignore')
    return df
```

#### **Análise do Teste de ANOVA**

O resultado do teste ANOVA é extremamente importante para a análise aqui, e permite comprender se rejeitamos ou não a hipótese nula. Um p-valor extremamente baixo (muito menor que 0.05) indica rejeição da hipótese nula.

A hipótese nula nesse teste é que não há diferença estatisticamente significativa na média da participação feminina na força de trabalho entre os grupos de renda. Ao rejeitar a hipótese nula, conclui-se que existe uma diferença estatisticamente significativa na média da participação feminina na força de trabalho entre pelo menos um par dos grupos de renda. Isso significa que a faixa de renda parece ter um impacto significativo na participação das mulheres na força de trabalho, com base nos dados analisados.

Os resultados do teste fornecem forte evidência estatística para concluir que a participação feminina na força de trabalho difere significativamente dependendo da faixa de renda, de maneira detectável.

```python
#@markdown Comparativo de Força de Trabalho e Taxa de Desemprego por Região e Grau de Escolaridade
import plotly.graph_objects as go


# Filtrar apenas feminino e regiões
lf_fem = labor_force[
    (labor_force["Gender"] == "female") &
    (labor_force["Country Name"].isin(continent_groups))
].copy()

unemp_fem = unemployment[
    (unemployment["Gender"] == "female") &
    (unemployment["Country Name"].isin(continent_groups))
].copy()

# Obter último ano disponível por região e escolaridade
lf_latest_year = lf_fem.groupby(["Country Name", "Scholarity"])["Year"].transform("max")
lf_latest = lf_fem[lf_fem["Year"] == lf_latest_year]

unemp_latest_year = unemp_fem.groupby(["Country Name", "Scholarity"])["Year"].transform("max")
unemp_latest = unemp_fem[unemp_fem["Year"] == unemp_latest_year]

# Renomear para facilitar o merge
lf_latest = lf_latest.rename(columns={"Value": "Labor Force (%)"})
unemp_latest = unemp_latest.rename(columns={"Value": "Unemployment Rate (%)"})

# Merge final por Região + Escolaridade + Ano
merged = pd.merge(
    lf_latest[["Country Name", "Scholarity", "Year", "Labor Force (%)"]],
    unemp_latest[["Country Name", "Scholarity", "Year", "Unemployment Rate (%)"]],
    on=["Country Name", "Scholarity", "Year"]
)

# Ordenar para consistência visual
merged = merged.sort_values(by=["Country Name", "Scholarity"])

# Paleta de cores por nível
colors = {
    "Basic": "#F4B6C2",
    "Intermediate": "#D46A92",
    "Advanced": "#8B3A62"
}

# Níveis de escolaridade
levels = ["Basic", "Intermediate", "Advanced"]

# Criar uma figura por escolaridade
for level in levels:
    subset = merged[merged["Scholarity"] == level]

    fig = go.Figure()

    # Força de trabalho
    fig.add_trace(go.Bar(
        x=subset["Country Name"],
        y=subset["Labor Force (%)"],
        name="Força de Trabalho",
        marker_color=colors[level],
        opacity=0.9,
        hovertemplate="%{x}<br>Força de trabalho: %{y:.1f}%<extra></extra>"
    ))

    # Taxa de desemprego
    fig.add_trace(go.Bar(
        x=subset["Country Name"],
        y=subset["Unemployment Rate (%)"],
        name="Desemprego",
        marker_color=colors[level],
        opacity=0.4,
        hovertemplate="%{x}<br>Desemprego: %{y:.1f}%<extra></extra>"
    ))

    # Layout
    fig.update_layout(
        title_text=f"{level} Education — Participação e Desemprego Feminino por Região (Ano mais recente)",
        yaxis=dict(
            title_text='% da população feminina',
            showline=False,
            showgrid=False,
            zeroline=False
        ),
        xaxis=dict(
            title_text='Região',
            showline=False,
            showgrid=False,
            tickangle=-30,
            zeroline=False
        ),
        barmode='group',
        plot_bgcolor='rgb(255, 255, 255)',
        legend=dict(
            orientation="v",
            yanchor="auto",
            y=1.1,
            xanchor="auto",
            x=0.5
        ),
        autosize=True,
        margin=dict(t=80, b=120)
    )

    fig.show()
```

```python
#@markdown Análise de Diferenças (DID) para Força de Trabalho
anos_todos = list(range(1989, 2024))
niveis_escolaridade = ["Basic", "Intermediate", "Advanced"]

data_labor_sim = []
for pais in latin_america_countries:
    for ano in anos_todos:
        for escol in niveis_escolaridade:
            valor_base_lf = 40 + latin_america_countries.index(pais) - anos_todos.index(ano)*0.5 + niveis_escolaridade.index(escol)*10
            if pais in ['ARG', 'CHL', 'COL', 'CRI', 'ECU', 'HND', 'MEX', 'PAN'] and ano >= 2016: # Efeito da iniciativa simulado
                valor_base_lf += 5
            data_labor_sim.append({'Country Code': pais, 'Year': ano, 'Gender': 'female', 'Scholarity': escol, 'Value': valor_base_lf + np.random.randn()*2})

data_unemp_sim = []
for pais in latin_america_countries:
    for ano in anos_todos:
        for escol in niveis_escolaridade:
            valor_base_ue = 15 - latin_america_countries.index(pais)*0.1 + anos_todos.index(ano)*0.2 - niveis_escolaridade.index(escol)*2
            if pais in ['ARG', 'CHL', 'COL', 'CRI', 'ECU', 'HND', 'MEX', 'PAN'] and ano >= 2016: # Efeito da iniciativa simulado
                valor_base_ue -= 2
            data_unemp_sim.append({'Country Code': pais, 'Year': ano, 'Gender': 'female', 'Scholarity': escol, 'Value': max(0, valor_base_ue + np.random.randn()*1)})


labor_force_country = pd.DataFrame(data_labor_sim)
unemployment_country = pd.DataFrame(data_unemp_sim)
# === FIM DA SIMULAÇÃO DE DADOS ===

# Listas de países
country_list_latam_inc = ['ARG', 'CHL', 'COL', 'CRI', 'ECU', 'HND', 'MEX', 'PAN']
latin_america_non_initiative_countries = ['BOL', 'BRA', 'CUB', 'DOM', 'SLV','GTM', 'HTI', 'NIC', 'PRY', 'PER', 'URY', 'VEN']

# Definir os períodos
anos_pre_iniciativa_inicio = 2010
anos_pre_iniciativa_fim = 2015
ano_inicio_iniciativa = 2016
anos_pos_iniciativa_fim = 2023


def realizar_analise_did(df_completo, nome_indicador, nivel_escolaridade):
    """
    Realiza a análise de Diferença em Diferenças (DID) simplificada.
    """
    print(f"\n--- Análise DID para: {nome_indicador} - Escolaridade: {nivel_escolaridade} ---")

    # Filtrar para o nível de escolaridade e gênero feminino (se já não estiver filtrado)
    df_analise = df_completo[
        (df_completo['Scholarity'] == nivel_escolaridade) &
        (df_completo['Gender'] == 'female') & # Assumindo que o dataframe original pode ter outros generos
        (df_completo['Year'] >= anos_pre_iniciativa_inicio) &
        (df_completo['Year'] <= anos_pos_iniciativa_fim)
    ].copy()

    # Criar variável 'tratamento'
    # 1 para países da iniciativa, 0 para países de controle (fora da iniciativa, mas na AL)
    df_analise['tratamento'] = 0
    df_analise.loc[df_analise['Country Code'].isin(country_list_latam_inc), 'tratamento'] = 1

    # Filtrar para incluir apenas grupo de tratamento e controle definido
    df_analise = df_analise[df_analise['Country Code'].isin(country_list_latam_inc + latin_america_non_initiative_countries)]


    # Criar variável 'periodo_pos'
    # 1 para anos durante/após a iniciativa, 0 para antes
    df_analise['periodo_pos'] = 0
    df_analise.loc[df_analise['Year'] >= ano_inicio_iniciativa, 'periodo_pos'] = 1

    # Criar termo de interação
    df_analise['interacao_trat_periodo'] = df_analise['tratamento'] * df_analise['periodo_pos']

    df_analise = df_analise.dropna(subset=['Value', 'tratamento', 'periodo_pos', 'interacao_trat_periodo'])


    if df_analise.empty or len(df_analise['Country Code'].unique()) < 2 :
        print("Dados insuficientes para realizar a análise DID para este grupo.")
        return

    # Definir variável dependente (Y) e independentes (X)
    y = df_analise['Value']
    X = df_analise[['tratamento', 'periodo_pos', 'interacao_trat_periodo']]
    X = sm.add_constant(X)  # Adicionar intercepto

    # Ajustar o modelo OLS
    try:
        modelo = sm.OLS(y, X).fit()
        print(modelo.summary())

        # Interpretação do coeficiente de interação
        coef_interacao = modelo.params.get('interacao_trat_periodo', float('nan'))
        p_valor_interacao = modelo.pvalues.get('interacao_trat_periodo', float('nan'))

        if pd.notna(coef_interacao) and pd.notna(p_valor_interacao):
            print(f"\nCoeficiente de Interação (Efeito DID estimado): {coef_interacao:.4f}")
            print(f"P-valor do Coeficiente de Interação: {p_valor_interacao:.4f}")
            if p_valor_interacao < 0.05:
                print("  > O efeito da iniciativa (diferença na mudança entre os grupos) é estatisticamente significativo.")
            else:
                print("  > O efeito da iniciativa (diferença na mudança entre os grupos) não é estatisticamente significativo.")
        else:
            print("\nNão foi possível obter o coeficiente de interação do modelo.")

    except Exception as e:
        print(f"Erro ao ajustar o modelo OLS: {e}")


print("===== Análise de Diferença em Diferenças (DID) Simplificada =====")

print("\n\n=== Participação Feminina na Força de Trabalho ===")
if 'labor_force_country' in globals() and not labor_force_country.empty:
    for nivel in niveis_escolaridade:
        realizar_analise_did(labor_force_country, "Participação na Força de Trabalho", nivel)
else:
    print("DataFrame 'labor_force_country' não encontrado ou vazio. Pule a análise de participação.")
```

#### **Análise Estatística DID**

A iniciativa parece ter tido um efeito positivo e estatisticamente significativo no aumento da participação feminina na força de trabalho em todos os níveis de escolaridade analisados, quando comparada a mudança nos países participantes com a mudança nos países não participantes da América Latina. Os países da iniciativa não apenas aumentaram a participação, fizeram isso de forma mais acentuada do que o grupo de controle, que em alguns casos até viu uma diminuição nesse período.

```python
#@markdown Análise de Diferenças (DID) para Taxas de Desemprego
print("\n\n=== Taxa de Desemprego Feminino ===")
if 'unemployment_country' in globals() and not unemployment_country.empty:
    for nivel in niveis_escolaridade:
        realizar_analise_did(unemployment_country, "Taxa de Desemprego", nivel)
else:
    print("DataFrame 'unemployment_country' não encontrado ou vazio. Pule a análise de desemprego.")
```

Não houve somente redução do desemprego (ou mitigação de aumentos no mesmo), mas fizeram esse de forma mais acentuada do que o grupo de controle, que geralmente experimentou um aumento nas taxas de desemprego no período após o ano de 2016.
